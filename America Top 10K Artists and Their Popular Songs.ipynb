```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
```

```python
#loading the dataset
artist_df = pd.read_csv('Artists.csv')
artist_us_songs_df = pd.read_csv('Top_Songs_US.csv')
```

# Artist Data Analysis (artist_df):

# 1-Gender Distribution:

```python
#Visualize the distribution of artists by gender using a count plot.
sns.countplot(x='Gender', data=artist_df)
plt.title('Gender Distribution of Artists')
plt.show()
```

# 2-Age Distribution:

```python
#Plot a histogram to show the distribution of artists' ages.
sns.histplot(x='Age', data=artist_df, bins=20, kde=True)
plt.title('Age Distribution of Artists')
plt.show()
```

# 3-Country-wise Analysis:

```python
#Explore the distribution of artists across different countries.
plt.figure(figsize=(25, 18))  # Adjust the width and height as needed
sns.countplot(x='Country', data=artist_df)
plt.title('Distribution of Artists by Country')
plt.xticks(rotation=0, ha='right')  # Adjust the rotation and alignment of x-axis labels
plt.show()
```

# 4-Popularity vs. Followers:

```python
#Scatter plot to visualize the relationship between popularity and followers.
sns.scatterplot(x='Popularity', y='Followers', data=artist_df)
plt.title('Popularity vs. Followers')
plt.show()
```

# Top Songs Data Analysis (artist_us_songs_df):

# 1-Album Type Distribution:

```python
#Check the distribution of album types.
sns.countplot(x='Album Type', data=artist_us_songs_df)
plt.title('Distribution of Album Types')
plt.show()
```

# 2-Explicit Content Analysis:

```python
#Visualize the distribution of explicit and non-explicit songs.
sns.countplot(x='Is Explicit', data=artist_us_songs_df)
plt.title('Explicit vs. Non-Explicit Songs')
plt.show()
```

# 3- Release Date Analysis:

```python
#Analyze the distribution of songs over time.
plt.figure(figsize=(16, 15))  # Adjust the width and height as needed
artist_us_songs_df['Release Date'] = pd.to_datetime(artist_us_songs_df['Release Date'])
artist_us_songs_df['Release Year'] = artist_us_songs_df['Release Date'].dt.year
sns.countplot(x='Release Year', data=artist_us_songs_df)
plt.title('Number of Songs Released Each Year')
plt.xticks(rotation=45)
plt.show()
```

# 4-Popularity vs. Song Duration:

```python
#Scatter plot to visualize the relationship between song popularity and duration.
sns.scatterplot(x='Popularity', y='Song Duration', data=artist_us_songs_df)
plt.title('Popularity vs. Song Duration')
plt.show()
```

# Time Series Analysis (Top Songs Data - artist_us_songs_df):

```python
# Convert 'Release Date' to datetime format
artist_us_songs_df['Release Date'] = pd.to_datetime(artist_us_songs_df['Release Date'])

# Extract year and month from 'Release Date'
artist_us_songs_df['Release Year'] = artist_us_songs_df['Release Date'].dt.year
artist_us_songs_df['Release Month'] = artist_us_songs_df['Release Date'].dt.month

# Group by year and count the number of songs released each year
songs_per_year = artist_us_songs_df.groupby('Release Year').size()

# Create a time series plot
plt.figure(figsize=(12, 6))
sns.lineplot(x=songs_per_year.index, y=songs_per_year.values, marker='o')
plt.title('Number of Songs Released Each Year')
plt.xlabel('Year')
plt.ylabel('Number of Songs')
plt.show()
```
